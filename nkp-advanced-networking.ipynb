###### 1. Multilayer Switches
***

In this presentation, we'll take a look at multilayer switches, which are network devices [Multilayer Switches.] that are capable of operating at both Layers 2 and 3 of the OSI model. 

So we've discussed earlier how switches reside at Layer 2 and base their packet forwarding decisions on the MAC address. 

Routers reside at Layer 3 and base their packet forwarding decisions on IP address. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Switching and Routing - Application Specific Integrated Circuit (ASIC)
***

Well, now we have a device that can, in fact, perform both basic functions, switching and routing. Now it uses what's known as an application specific integrated circuit, which is also referred to as an ASIC circuit, to accomplish the routing tasks. 

So, again, it is just a 2-in-1 type of device. So in terms of the actual network configuration, we see here [In the diagram there are two networks: Network A and Network B. Each network has four systems that are interconnected with a switch and both the networks are connected with a Multilayer Switch. ] Network A and of course Network B, and you might have traditionally placed a router in between them, provided they were using separate network addresses. So, again, perhaps Network A was 192.168.0 and network B was 192.168.1. Those are separate network addresses. 

Well, a multilayer switch can still route between them. It can examine the IP address and decide whether or not it should forward the packet. But it can also just act as a switch. So in terms of reconfiguring, you wouldn't have to do anything. But in fact, you could if you wanted to. You could make Network A, Segment A, and make it 192.168.0, and then maybe the first 5 addresses or 10 addresses or 50 addresses, it doesn't matter within that subnet. 

But then network B could become just Segment B, and it could still be 192.168.0 as long as the address range did not overlap. So, again, if the first one went to, let's say 10, then the second one could start at 11. So as long as there were no conflicts, then you could make this entire configuration one network. But, again, I stress, you didn't have to, that was just an option. You could leave them exactly as they were, because a multilayer switch is able to assess the forwarding determination based on the IP address as well.

You have to program routers to say that this is the network on this side and this is the network on the other side. But because they are able to detect the addresses now at Layer 3, they can do, effectively, the same thing, and they can build tables based on the IP address. So there is an autoconfiguration with respect to both switching and routing. So there's no requirement to learn a new kind of IP switching technology, it configures itself.

###### 3. Load Balancers
***

Now in this presentation, we'll introduce the idea of load balancing. And in particular, we'll look at the hardware load-balancing device, or HLD. [Load Balancer.] 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Hardware Load-Balancing (HLD)
***

Now there are software mechanisms in place that can also load balance. But, for the time being, we'll talk about the hardware device. 

This is a physical networking appliance that is used to distribute traffic across multiple servers as evenly as possible. Now a key component is that those servers all have to be performing the same task. Now they don't necessarily have to be dedicated to that task. But every server that is behind the load balancer must at least supply the same consistent service as every other system. 

So a good example is a website. If I have a very busy website, then one server might not be able to handle the load. So I think to myself, well, I'll create a server farm and I'll have the site hosted on two servers, then either one can respond. That's the basic idea, but then you also want to ensure that server one is not handling, let's say, 90% of the load while server two is only handling 10. Now each of those servers could then also perform other tasks that don't necessarily need to be load balanced. And that's what I referred to when I said they have to at least have one consistent service. But, typically, that's what you are concerned with. Let's make sure that we can handle the load of the website traffic. But let's spread it across multiple servers. And, by no means, does it need to be only two servers.  You could have dozens upon dozens as long as they are all hosting that same website or performing the same service. 

Now in terms of routing the traffic, it will try its best to distribute the traffic evenly. But evenly doesn't just come down to the numbers. So using a simple example, if it was only two servers then you might think well, let's split it up 50/50. But you also have to consider availability, resource utilization, the number of connections overall, and quite simply, the performance of that server. 

So if, for example, server one is a very good, very robust server, while server two is not so good, well, then, in fact, you would want server one to handle more of the requests. So it will take into account all of these considerations. And particularly once you start dealing with, let's say, ten servers, then availability also comes into play. Because one of them might go down. Two of them might go down. In which case, the load balancer will ensure that it does not forward the requests to a server that has gone down. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Benefits
***

So in terms of benefits, it minimizes the probability that servers will become overwhelmed by the traffic. So, again, a slight exaggeration that I mentioned earlier, but if one server was handling 90% and another was only handling 10, well, you're certainly more likely to see server one become overwhelmed, it's doing too much. 

It can also optimize the bandwidth to each computer or terminal. Again, it can take into account a number of considerations. So it can forward a request off to a server that is simply not utilizing the bandwidth as much as it should. 

And it can minimize the network downtime. Because if you do have multiple servers, even just two, and one of them goes down, there is still another one to service the request. So you have less downtime, better resiliency. 

Now you can also perform traffic prioritization and end-to-end application monitoring. Which involves, in essence, saying, if a request comes in for this type of service and that is deemed to be a little more important than a different type of service, then we can give that a little more priority. Because you don't have to balance just one type of service. You can balance any number of services. So that allows you to put more weight on the type of protocol or the type of request that is coming in. And that monitoring ability is to flat out determine which systems have been doing what. So we can ensure that the balance is there. That we aren't seeing overutilization and underutilization. 

It can also provide user authentication with the load balancing mechanisms. 

And you can protect against malicious activity such as a denial-of-service attack. Because clients that are in front of the load balancer, if you will, do not see behind it. So launching an attack against this service has to go through the load balancer. So it can be configured, of course, to reject those types of requests that don't seem to be legitimate. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Multiple Servers handling a Single Service
***

So, ultimately, if you do require this type of configuration where you want to have multiple servers handling a single service, then it typically makes sense to implement the load balancer to ensure the optimal performance of that service.

###### 4. Intrusion Detection System (IDS) and Intrusion Prevention System (IPS)
***

In this presentation, we'll take a look at network intrusion from two different perspectives, [Intrusion Detection System.] one of detection and one of prevention. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Intrusion Detection System (IDS)
***

So beginning with detection, the idea really is that key word, it detects suspicious activity on either an individual host or the entire network. 

It is a system that's designed to analyze traffic patterns and to, effectively, try to determine what is normal, for lack of a better word. Then anything that occurs that is not normal, or out of the ordinary, might be considered to be an intrusion. But, in terms of mitigation, it can assist in mitigating that detected activity, but typically in the form of just logging it. Taking note of the fact that something did occur. [IDS.] So what it really comes down to is that the IDS is the solution that can detect suspicious activity, but it does not necessarily stop it. It is more so up to you to go through the logs and assess what did take place and then, perhaps, take some action based on what you are seeing. So it really is the key word of intrusion detection. Detecting something does not prevent it. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Intrusion Prevention System (IPS)
***

[IPS.] So this is where we see the Intrusion Prevention System come into play. And this is a solution that can both detect the suspicious network activity, and furthermore, in many cases, prevent an attack.

So what you typically find is that the IDS will perform the detection, learn what is normal and what is abnormal. And then based on that information, you can then configure the IPS to flat out stop that type of traffic. You can effectively say that yes, this appears to be an intrusion. So if you see that same pattern again, stop it from happening. 

So, typically, most of the devices can perform both tasks, but there needs to be a little bit of a learning curve, if you will.

< [Table of Contents](#top) | [References](#references) >

###### Network Firewalls
***

[Network Firewalls.] Now, I also want to include the network firewall in this presentation because it's similar, but it's not really the same. 

The firewall is simply configured to allow or deny network traffic based on packet header contents such as the IP address, a protocol type and/or a port number. 

And it's placed, very much like an IDS/IPS, in between the Internet and your private network. 

But the firewall doesn't really assess anything other than what you see here. And it doesn't really make any decisions based on previous historic patterns or anything along those lines. If the IP address appears to be correct, if the protocol type appears to be correct, and the port number appears to be correct, well, then the firewall will let that traffic through. 

But, of course, a skilled attacker will be able to find their way around that, if you will. They disguise the packet so that they appear to be legitimate. They have the correct IP address, they have the correct protocol type, and the correct port number. It is simply the contents that, of course, are malicious. So it's never a bad idea to have both a firewall and an IDS/IPS for the highest levels of protection. But firewalls in and of themselves only examine those contents and, effectively, allow or deny the traffic accordingly.

###### 5. Proxy Servers
***

[Proxy Servers.] Are implemented to eliminate direct connectivity between your internal clients and Internet resources. 

And in fact, in the opposite direction as well, Internet clients and internal resources. Now, that's actually what's known as a reverse proxy configuration.

But it protects the true host identity because all outgoing traffic appears to come from the proxy. And in fact, it's not just a matter of it appearing to come from the proxy, it does come from the proxy. So we see the internal clients using private IP addresses. They just simply want to go to a website, for example. The proxy server intervenes, if you will, and says hold on, I'll get that site for you. So it has a public IP address configured that faces the Internet. It is the proxy server that goes out and retrieves the site and then it sends it back to the client.

So again, you get this protection of the internal clients because it's only ever the proxy server that directly accesses the Internet. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Configuration
***

Now, it has to be configured to proxy a specific application. So for example, if you are wanting to gain access to a website, it would have to proxy the HTTP protocol. 

And if you also wanted it to proxy FTP requests, you certainly could. But you do have to tell it which application protocols are going to be proxied. Anything that is not listed, in essence, would bypass the proxy. 

Now, transparent proxies do not really require any kind of client configuration. In other words, you don't have to do a whole lot on the client to configure it. You just really tell it to go through the proxy, that's about it. And a lot of browsers these days are capable of detecting a proxy as well. So from the client's perspective, it is all very transparent. You might not even know that you are going through a proxy server. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Benefit - Cached Content
***

So again, the idea is that the proxy server is actually the system that retrieves the Internet content for the internal requesting client. But, another nice feature is that the content that does come back can be cached to speed up subsequent requests. In other words, if I request website X, then the server can cache some of that content. Now, these days we know a lot of websites are very dynamic. But some of it can still be cached. So that if you go and request the same site, a good portion of it can actually come back from the proxy itself. And it doesn't even have to go out to the Internet. So you get your site back at LAN speeds, not Internet speeds. So they can dramatically increase the performance as well. And again, the internal client identity is never revealed, so you get a lot of benefits from using proxy servers.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Benefit - Control or Restriction or Security
***

Now, another benefit is that you can implement a certain degree of control or restriction or even security. Because proxy servers can examine the packet headers and the payload. So for example, you can block connections to things like social networking sites outside of lunch hours. Or you can flat out block an entire site based on the domain name, for example. 

So you can ensure that your people aren't wasting time during business hours. You can control which sites are accessible and which ones are blocked or restricted. And you can even implement these time controls, so that maybe it is allowed during certain hours. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Caching Servers - Time To Live (TTL)
***

Now, as mentioned earlier, proxy servers can cache the content. So as such, they're often referred to as caching servers. Now this depends on how the proxy server is configured because you don't have to enable caching. That is really up to you. But if you do enable it, then you might also want to consider cache aging timers, also known as the time to live, or TTL, value. In other words, how long is the cached content valid for? As mentioned, a lot of sites these days are very dynamic. They change all the time, especially weather or news sites. So in that event, if I go to the site and then it is cached. And then you go to the site later on, the decision is well, is that cached content still valid or has it changed? So if you feel that you are always accessing, or at least frequently accessing, very dynamic sites, then you should have a very short TTL. Because the original content is changing. But if you feel that you access fairly static sites more often, then you can increase the TTL and you'll get better performance. But once the TTL expires, the cached content is removed. So again, if it is very dynamic, cache it for maybe 15 minutes or something like that. Then it expires, then the next person to go to that site will have to go all the way to the originating site to get the current content. 

So it's a little bit slower but it's more up to date. Cached is faster but might not be as current.

< [Table of Contents](#top) | [References](#references) >

***
###### Prefetched Internet Content
***

Now there's also prefetched Internet content, which can be scheduled. This can speed up content access when requested by internal clients. 

Now again, you have to consider the static versus dynamic nature of the cached content. So again, this is quite literally proxy servers realizing that these sites are very often accessed. Therefore, we can proactively request that content and cache it ahead of time. 

Now again, this is something that is entirely up to you. You can enable this feature, disable it, configure it however you want. So it does depend, to a degree, on knowing what the normal behavior is within your environment and which sites are accessed most frequently. But you can, of course, examine that information and decide for yourself over time if these features are beneficial.

***
###### Reverse Proxy Servers
***

[Reverse Proxy Servers.] Now, the reverse proxy was touched on earlier. And this is effectively the opposite direction, the requesting client is on the Internet. 

So they are trying to gain access to resources within your environment, for lack of a better word. And we see here the requesting client coming in through the reverse proxy to access a web server. 

So if you think about the regular proxy representing the clients, then this is the exact opposite. The reverse proxy represents the servers. 

Now, the idea is that you might have multiple web servers in a lot of cases. So if a request comes in for a particular site, the proxy server is also responsible for directing that client to the appropriate web server. So if I have even just two web servers, let's say I have Site A on Server A and Site B on Server B. 

Then it's the job of the reverse proxy to forward me off to the correct web server. So it listens for these connections on the appropriate ports. And then those incoming connections are forwarded to the appropriate internal host. [For example, it listens for TCP port 80 web site connections.] So that's just the reverse nature of this type of request. It originates from the Internet, not from the internal environment. 

Now these are normally placed in a demilitarized zone for publicly accessible services. This still protects the true identity of the internal host. But now the internal host is the server, not the client. And again, it's transparent to the external users. The use of a reverse proxy is not known to the clients. They have no idea that they aren't directly communicating with the web server. But that's the idea, I want all incoming requests to hit the reverse proxy. So this protects all of your servers that are behind the proxy and allows all traffic to be controlled through the proxy.

###### 6. VPN Concentrators
***

All right, in this presentation, we'll take a look at the VPN concentrator, [VPN Concentrator.] which is a networking device used to provide remote access for virtual private networks. 

Now, just to touch on the VPN itself for a moment, this is implemented when you have disconnected workers. Those who work from home or those who travel and need to access the internal resources of your organization over the Internet. 

So once a VPN is established, they can access those resources, such as the printers, the servers, the storage, just like they were in the actual internal office. So, typically, what you'll see is software on those computers for the disconnected users that they will launch. And they will provide the necessary connectivity information, such as the username and password and the server or address to which they want to connect. Then once they establish that connection, they might as well be in the office. They can work just like anyone else, but, of course, from the outside.

But the VPN concentrator is implemented to act as a dedicated device. Because most network servers have the ability to create and establish VPN connections. But then you have to take one of your servers and, basically, expose it to the Internet because the system to which the users want to connect needs to be accessible from the Internet. 

So by using a VPN concentrator, you can keep all of your servers behind the concentrator and not directly exposed to the Internet. 

Plus, as a dedicated device, the VPN concentrator allows typically higher throughput and other benefits, such as handling the encryption. Because, of course, we are using the public Internet as our means to communicate. So we want to ensure that the data is encrypted.

So a site-to-site maintains a much more persistent connection that all users can use, rather than individuals making multiple connections all of the time. That more casual type of connection, such as the travelling user who only needs access periodically while they are travelling, is known as a point-to-point. That's when they open the software, they enter in the information, they establish a connection, they go about their business and then they close it down once they are finished. ###### Protocols - Remote Authentication Dial-In User Service (RADIUS)
***

[RADIUS.] So beginning with RADIUS, this is the Remote Authentication Dial-In User Service. 

It's an 802.1x standard based AAA protocol for remote connections, but typically involves centralizing the authentication request. 

If you have just a single network access server, then you can configure the authentication and the security on that server. And since it's the only one, it handles all connections. So you know that every one is going to be configured the same. 

But in larger, more robust environments, you might have multiple network access servers. In which case, you might find yourself configuring the exact same security configuration and authentication configuration on each of those servers. So, in other words, the policy is the same on every one of those servers and as this grows, if you have 5 or 10 or 20, you end up configuring the exact same security policy across all of those servers.

RADIUS says, well, only implement a single security policy on the RADIUS server. Then when the requests come in from the network access servers, they will simply forward the request back to the RADIUS server. It will handle the authentication and then allow the user through. 

So this way, you only configure a single policy as opposed to several policies which are effectively identical. 

So this is very widely used for VPNs, various types of access points, and any other remote access type with respect to the security. It just allows you to simplify the process by implementing that single consistent policy and then every access point adheres to that same policy.

But, again, it is a Cisco developed protocol so it's not only supported by Cisco devices, it is an open standard. But, typically, you will see it implemented when you do have Cisco hardware.

###### 8. Unified Threat Management (UTM) Appliances
***

In this presentation, we'll take a look at Unified Threat Management or UTM. [UTM.]

This is an all-in-one security device that was designed to replace the typical firewall. Because a firewall, in and of itself, is only designed to protect the internal environment based on things like the protocol and the port and the IP address. That's not particularly comprehensive. There are several means by which you can establish a connection including routing and VPNs. But there are a number of different threats to those types of connections. Anti-virus, anti-spyware, email, and anti-spam, intrusion prevention systems, quality of service and bandwidth management, web security. All of those things have to be considered. And the UTM was designed to do exactly that. 

To take into consideration all of the methods of connectivity, and to implement an appropriate type of security for each of those types of connections. 

So it consolidates many security services into the one device. Whereas in the past, you would have purchased many devices or applications to handle each one of those specific threats.

The UTM secures your network by providing many standard services such as a firewall, intrusion detection and prevention, antivirus and anti-spyware, VPN connectivity and load balancing, and content filtering, such as on a proxy server to implement web security in many cases. 

But the enterprise versions might also include features such as identity based access control, quality of service, SSL and SSH inspection for security, and network intrusion prevention as well. 

Now there are always considerations. The principal advantage of a UTM is that it can certainly reduce the complexity. Everything is configured within the one device. But at the same time, this introduces the principal disadvantage of it becoming a single point of failure. If the UTM goes down, all of the services it supplies go down with it. Now, you might still retain some of the other services as backups, or maybe you even have a standby UTM. But it's always something to consider. How will you address the availability of the UTM? But it certainly does reduce the complexity because everything is centralized. And you don't have these different servers, and different applications and different licensing concerns to address each of those threats, hence the term Unified Threat Management.
